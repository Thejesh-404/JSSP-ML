```python
import numpy as np
import torch

# Load the pickle file
with open('data_zhang_L2D/6_6_Seed200.pkl', 'rb') as f:
    data = pickle.load(f)

# Modify each tensor in the list and concatenate along a new axis
modified_data = []
for tensor in data:
    new_column = tensor[:, 3] / 10  # Calculate the new column as 1/10 of the 4th column
    new_column = np.expand_dims(new_column, axis=1)  # Expand dimensions to make it a column vector
    modified_tensor = np.concatenate((tensor, new_column), axis=1)  # Concatenate the new column
    modified_data.append(modified_tensor)

# Stack modified tensors along a new axis to create a single tensor
modified_data_tensor = torch.tensor(np.stack(modified_data, axis=2))

# Save the modified tensor to a new pickle file
with open('data_zhang_L2D/modified_data.pkl', 'wb') as f:
    pickle.dump(modified_data_tensor, f)
```

```python
file_path = 'data_zhang_L2D/modified_data.pkl'

with open(file_path, 'rb') as file:
    d = pickle.load(file)

print(d.size())
```
